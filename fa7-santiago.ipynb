# (i) recursion

```python
%%time
def countdown(n):
    if n <= 0:
        print('Blastoff!')
    else:
        print(n)
        countdown(n-1)

n = 2962
countdown(n)
```

# (ii) using while statement

```python
%%time
def countdown(n):
    while n > 0:
        print(n)
        n = n - 1
    print('Blastoff!')
    
n = 2962
countdown(n)
```

**Using while statement is faster. For each n, the difference is seen when rendering 10000. The limit for the recursion is 2962.**
